# Moving Average (MA) Model

```python
import pandas as pd

sp500 = pd.read_csv("./S&P500.csv")
sp500
```

## Simple Moving Average

$$
\frac{a_1,a_2,a_3, ..., a_n}{n} 
$$

```python
# Average over 3 Months
sp500["SP500_SMA_3"] = sp500.iloc[:,1].rolling(window=3).mean()
sp500[["Date", "SP500", "SP500_SMA_3"]]
```

```python
# Average of 30 Months 
sp500["SP500_SMA_30"] = sp500.iloc[:,1].rolling(window=30).mean()
sp500[["Date", "SP500", "SP500_SMA_3","SP500_SMA_30"]]
```

## Exponential Moving Average

EMA = Closing Price x multiplier + EMA(Previous Day) x (1 - multiplier)

- Pandas has the `ewm()` function for EMA
- The span, $\alpha$, formula is the formula for the multiplier when the smoothing factor is 2
    
$$
\alpha = \frac{2}{observations + 1} 
$$
    
- When `adjust = False`, the exponentially weighted function is calculated recursively
    
$$
y_0 = x_0 
$$
$$
y_t = \alpha x_t + y_{t-1} (1 - \alpha)
$$

```python
sp500["SP500_EMA_30"] = sp500.iloc[:,1].ewm(span=30, adjust=False).mean()

multiplier = 2 / (30 + 1)
sp500.loc[0, "SP500_EMA_30_Man"] = sp500.loc[0,"SP500"]

for i in range(1, len(sp500)):
    sp500.loc[i, "SP500_EMA_30_Man"] = (sp500.loc[i,"SP500"] * multiplier) + (sp500.loc[i - 1,"SP500_EMA_30_Man"] * (1 - multiplier)) 

sp500[["Date", "SP500", "SP500_SMA_3","SP500_SMA_30", "SP500_EMA_30", "SP500_EMA_30_Man"]]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=[18,12])
plt.grid(True)
plt.plot(sp500["SP500"], label = "Original")
plt.plot(sp500["SP500_SMA_3"], label = "SP500_SMA_3")
plt.plot(sp500["SP500_SMA_30"], label = "SP500_SMA_30")
plt.plot(sp500["SP500_EMA_30"], label = "SP500_EMA_30")
plt.legend(loc=2)
```
